```python
def target_combination(req, option={}, power={}, num=math.inf):
    output = []

    if 'crit_' in req:
        req['critRate_'] = req['crit_']
        req['critDMG_'] = req['crit_']
        req.pop('crit_')

    if not option:
        option[None] = 0

    for opt in option.keys():
        if opt == 'crit_':
            base = {'critRate_': option['crit_'], 'critDMG_': option['crit_']}
        else:
            base = {opt: option[opt]}

        keys = list(power.keys())
        # Generate all possible lengths for combinations (0 to len(keys))
        for r in range(len(keys) + 1):
            # Generate all combinations of length r
            for subset in itertools.combinations(keys, r):
                temp = base.copy()
                temp.update(req)
                temp[opt] = option[opt]
                for i in subset:
                    if i == 'crit_':
                        temp['critRate_'] = power['crit_']
                        temp['critDMG_'] = power['crit_']
                    else:
                        temp[i] = power[i]
                output.append((temp, num))

    return output
```
